```python
#Daten laden
einstellungen.set_plt_options()
karte = karten.BasisKarte()

def berechne_entfernung(von, zu):
    '''
    Zur berechnung von Entfernungen
    '''
    a_q = abs(von[0]-zu[0])
    b_q = abs(von[1]-zu[1])
    return math.ceil(math.sqrt(a_q+b_q)) / float(2)
```

##Approach Greedy:
1. Waehle einen zufaelligen Termin
- Waehle einen Techniker mit den wenigsten Terminen, der in Fahrtweite ist (2 mal Fahrtweg + Dauer)
- Fuer jeden Tag ab min_planungshorizont pruefe, ob Max Arbeitszeit nicht ueberschritten werden wuerde, wenn vom vorherigen Termin direkt dorthin und danach nach hause gefahren wird
- Wenn Planungshorizont erreicht ist, nimm den naechstdichteren aus Gruppe 1
- Wenn keiner vorhanden ist, nimm den mit den naechstwenigen Terminen
- Wenn kein Techniker mehr verfuegbar ist, fuege Termin in Backlog hinzu und nimm einen neuen
- Wiederhole, bis kein Termin mehr vorhanden ist

Zu finden in $solver.py$

#Auswertung von Lösungen

```python
def zeit_gefahren_berechnen(p):
    '''
    berechnet fuer ein Problem fuer jeden Techniker
    die durchschnittliche Fahr- und Arbeitszeit
    '''
    zeitaufteilung = []
    for tech in p.techniker:
        zeit_gefahren = 0
        zeit_gearbeitet = 0
        anzahl_termine_tech = 0
        name = tech.name
        arbeitszeiten = []
        #iteriere ueber Tage
        for day, apts in tech.termine.kalender.iteritems(): 
            #wenn es an dem Tag Termine gibt
            if apts:
                orte = []
                for time, apt in apts.iteritems():
                    orte.append(apt.kunde.ort)
                    zeit_gearbeitet+=apt.dauer
                    anzahl_termine_tech +=1
                for i, o in enumerate(orte):
                    #erster ist von Heimatort
                    if i == 0:
                        zeit_gefahren += berechne_entfernung(tech.heimatort, o)
                    #ab zweiten von vorherigem
                    else:
                        zeit_gefahren += berechne_entfernung(orte[i-1], o)
                    #Nach dem letzten gehts nach hause
                    if i == len(orte) -1:
                        zeit_gefahren += berechne_entfernung(tech.heimatort, o)
                if day >= einstellungen.min_planungshorizont:
                    letzter_term = max(apts.keys())
                    zuhause = letzter_term + apts[letzter_term].dauer + berechne_entfernung(apts[letzter_term].kunde.ort, tech.heimatort)
                    arbeitszeiten.append(zuhause - einstellungen.zeiten_ab)
                     
        zeitaufteilung.append((name, 
                               anzahl_termine_tech, 
                               zeit_gefahren, 
                               zeit_gearbeitet, 
                               round(np.mean(arbeitszeiten), 2), 
                               round(np.std(arbeitszeiten),2)))
    return zeitaufteilung
zeit_gefahren_berechnen(testproblem)
```

```python
def berechne_arbeitszeiten(p):
    '''
    berechne die Arbeitszeit und Std fuer ein Problem fuer alle Techniker
    '''
    arbeitszeiten = []
    for tech in p.techniker:
        for day, apts in tech.termine.kalender.iteritems():  
            if day >= einstellungen.min_planungshorizont:
                if apts:
                    letzter_term = max(apts.keys())
                    zuhause = letzter_term + apts[letzter_term].dauer + berechne_entfernung(apts[letzter_term].kunde.ort, tech.heimatort)
                    arbeitszeiten.append(zuhause - einstellungen.zeiten_ab)
                else:
                    pass
    return round(np.mean(arbeitszeiten), 2), round(np.std(arbeitszeiten), 2)
berechne_arbeitszeiten(testproblem)
```

##Entwicklung eines Scores fuer einen Terminplan
1. Nur Anteil bewerten, der nicht im Backlog ist
- Strafe fuer verspaetete Termine (Termine die zu spaet anfangen)
- Strafe fuer Tage ueber 10 Stunden
- Strafe fuer Wochen mit ueber 40 Stunden

```python
#Testproblem zur Entwicklung
scoreprob = problem.Problem(1, karte, 5, 20, 200)
scoreprob, bl_ausw = solver.greedy_solver_erweitert(scoreprob)
```

```python
def get_fehler_fuer_techniker(techniker):
    '''
    Hilfsfunktion, gibt zurueck:
    Anzahl der Tage mit ueber 10 Stunden Arbeitszeit zaehlt
    Anzahl zu spaeter Termine
    Anzahl Wochen mit ueber 40 Stunden
    '''
    
    curr_plan = techniker.termine.kalender
    curr_ort = techniker.heimatort
    
    tage_ueber_max = 0
    verspaetungen = 0
    
    wochen_ueber_max = 0
    
    curr_wochentag = 0
    wochenarbeitszeit = 0
    #iteriere ueber alle Tage
    for tag, termine in curr_plan.iteritems():
        curr_wochentag += 1
        if termine:
            #test ob zu lange gearbeitet wurde
            letzter_termin = termine.keys()[-1]
            letzter_ort = termine[letzter_termin].kunde.ort
            
            arbeitszeit = letzter_termin + termine[letzter_termin].dauer + berechne_entfernung(letzter_ort, curr_ort)
            arbeitszeit -= einstellungen.zeiten_ab
            
            #fuer verspaetungen
            letzter_termin_ende = einstellungen.zeiten_ab
            letzter_termin_ort = techniker.heimatort
            curr_verspaetung = 0
            erster_termin = True
            
            #Benutze wirkliche Dauer, um Verspaetungen zu messen
            for stunde, app in termine.iteritems():
                arbeitszeit += app.wirkliche_dauer-app.dauer
                #Verspaetungen durch Termine vorher gehen nur bei Terminen ab dem 2.
                if not erster_termin:
                    fahrtzeit_zu_curr = berechne_entfernung(letzter_termin_ort, app.kunde.ort)
                    if stunde < letzter_termin_ende + fahrtzeit_zu_curr + curr_verspaetung:
                        curr_verspaetung = letzter_termin_ende + fahrtzeit_zu_curr - stunde
                        verspaetungen += 1
                    else:
                        curr_verspaetung = 0
                
                letzter_termin_ende = stunde + app.wirkliche_dauer
                letzter_termin_ort = app.kunde.ort
                erster_termin = False
                
            if arbeitszeit > einstellungen.max_arbeitsdauer:
                tage_ueber_max += 1
            
            wochenarbeitszeit += arbeitszeit
            if  curr_wochentag % 5 == 0:
                if wochenarbeitszeit > 40:
                    wochen_ueber_max += 1
                wochenarbeitszeit = 0
                curr_wochentag = 0
            
    return (tage_ueber_max, verspaetungen, wochen_ueber_max)


def calc_solution_score(testproblem, backlog):
    '''
    Berechnet eine Zahl, die die Qualitaet einer Loesung darstellt
    '''
    #Wichtungen
    bl_w = 20              # Fuer einen Termin
    tage_w = 0             # Platzhalter fuer Strafe fuer Tage ueber 10 h
    tage_pro = -70         # Strafe fuer Tag ueber 10 h
    versp_w = -5           # Strafe fuer Verspaeteten Termin
    woche_w = -70          # Strafe fuer Woche ueber 40 h
    
    solution_score = 0
   
    #Schritt 1 - Nur Anteil, der Scheduled werden konnte
    anteil_zu_spaet = float(len(backlog))/testproblem.anzahlTermine
    bl_w = testproblem.anzahlTermine*bl_w*(1-anteil_zu_spaet)
    #print "Anteil ohne Backlog:", bl_w
    
    verspaetungen = 0
    wochen_zu_lang = 0
    for t in testproblem.techniker:
        techniker_auswertung = get_fehler_fuer_techniker(t)
        tage_w += techniker_auswertung[0]*tage_pro
        verspaetungen += techniker_auswertung[1]
        wochen_zu_lang += techniker_auswertung[2]
        
    #print "Strafe fuer ueber 10:", tage_w
    
    versp_w = float(verspaetungen)*versp_w
    #print "Strafe fuer verspaetungen:", versp_w

    woche_w *= wochen_zu_lang
    #print "Strafe fuer Wochen ueber 40h:", woche_w
    
    anteile_gefahren = np.mean([float(t[3])/(t[2]+t[3]) for t in zeit_gefahren_berechnen(testproblem) if t[3] > 0])
    #print zeit_gefahren_berechnen(testproblem)
    
    solution_score += bl_w
    solution_score += tage_w
    solution_score += versp_w
    solution_score += woche_w
    solution_score = round(solution_score*anteile_gefahren)
    return int(solution_score)
```

```python
#Berechne die Lösung mit unterschiedlich vielen Terminen
scores_simpel = []
scores_erweitert = []
scores_final = []

for i in xrange(20, 350, 1):
    score_s = []
    score_e = []
    score_f = []
    for counter in xrange(5):
        auswertungsproblem = problem.Problem(1, karte, 5, 20, i)
        auswertungsproblem, backlog_auswertung = solver.greedy_solver_final(auswertungsproblem)
        score_f.append(calc_solution_score(auswertungsproblem, backlog_auswertung))

        auswertungsproblem.reset_problem()
        auswertungsproblem, backlog_auswertung = solver.greedy_solver_erweitert(auswertungsproblem)
        score_e.append(calc_solution_score(auswertungsproblem, backlog_auswertung))

        auswertungsproblem.reset_problem()
        auswertungsproblem, backlog_auswertung = solver.greedy_solver(auswertungsproblem)
        score_s.append(calc_solution_score(auswertungsproblem, backlog_auswertung))
        
    scores_simpel.append(np.mean(score_s))
    scores_erweitert.append(np.mean(score_e))
    scores_final.append(np.mean(score_f))
    
    
```

```python
#Hilfsfunktionen fuer Auswertungstabelle (siehe naechste Zelle)
def calc_closer(techniker1, techniker2):
    '''
    zum Berechnen, wie viele Termine nicht den optimalen Techniker haben
    '''
    ort_1 = techniker1.heimatort
    ort_2 = techniker2.heimatort
    orte_array = []
    for tag in [v for i, v in techniker1.termine.kalender.iteritems()]:
        for asd, term in tag.iteritems():
            orte_array.append(term.kunde.ort)
            
    counter_closer = 0
    for ort in orte_array:
        if berechne_entfernung(ort_1, ort) > berechne_entfernung(ort_2, ort):
            counter_closer+=1
    return counter_closer

def calc_closer_byhalf(techniker1, techniker2):
    '''
    Zum berechnen, wie viele Termine mehr als 30 minuten sparen koennen
    '''
    ort_1 = techniker1.heimatort
    ort_2 = techniker2.heimatort
    orte_array = []
    for tag in [v for i, v in techniker1.termine.kalender.iteritems()]:
        for asd, term in tag.iteritems():
            orte_array.append(term.kunde.ort)
            
    counter_closer = 0
    for ort in orte_array:
        if berechne_entfernung(ort_1, ort) > berechne_entfernung(ort_2, ort)+.5:
            counter_closer+=1
    return counter_closer
```

```python
def finde_termine_zum_verbessern(techniker1, testproblem):
    '''
    Finde Termine, die verbessert werden koennen
    '''
    ort_1 = techniker1.heimatort
    termine_array = []
    for i, tag in techniker1.termine.kalender.iteritems():
        for stunde, term in tag.iteritems():
            entf = berechne_entfernung(ort_1, term.kunde.ort)
            schon_gefunden = False
            info_array = {}

            for tech in testproblem.techniker:
                entf2 = berechne_entfernung(tech.heimatort, term.kunde.ort)
                if entf > entf2 + .5:
                    if not schon_gefunden:
                        schon_gefunden = True
                        info_array['entfernung_original'] = entf
                        info_array['entfernung_potential'] = entf2
                        #Es kann mehrere Techniker zum Tauschen geben
                        info_array['neuer_tech'] = [tech]
                        info_array['orig_tech'] = techniker1
                        info_array['orig_tagplan'] = tag
                        info_array['orig_tag'] = i
                        info_array['orig_stunde'] = stunde
                        info_array['termin'] = term
                        info_array['selber_tag'] =  [term1.kunde.name for term1 in tag.values()].count(term.kunde.name)
                    if schon_gefunden:
                        info_array['neuer_tech'].append(tech)
            if schon_gefunden: 
                termine_array.append(info_array)
    return termine_array
try:
    finde_termine_zum_verbessern(local_test.techniker[0], local_test)[-1]             
except:
    print "keine termine fuer ersten tech verbesserbar"
```

```python
#Filtert die Termine, die mehr als einen Termin bei dem Kunden an einem Tag haben
def count_direkt_verbesserbar(techniker1, testproblem):
    return len([i for i in finde_termine_zum_verbessern(techniker1, local_test) if i['selber_tag'] == 1])
def direkt_verbesserbar(techniker1, testproblem):
    return [i for i in finde_termine_zum_verbessern(techniker1, local_test) if i['selber_tag'] == 1]
for t in local_test.techniker:
    print t.name, count_direkt_verbesserbar(t, local_test)
```

```python
def finde_moeglichkeiten(techniker1, techniker2, termin):
    '''
    Hilfsfunktion, die Tauschkandidaten fuer einen Termin Identifiziert
    '''
    termine_array = []
    zu_sparen = termin['entfernung_original'] - termin['entfernung_potential']
    for i, tag in techniker1.termine.kalender.iteritems():
        for stunde, term in tag.iteritems():            
            entf1 = berechne_entfernung(techniker1.heimatort, term.kunde.ort)
            entf2 = berechne_entfernung(techniker2.heimatort, term.kunde.ort)
            #Wenn Einsparpotential groesser als Verlust, wenn man Termine tauscht, dann lohnt sich Tausch
            if zu_sparen > entf1 - entf2:
                #Nur, wenn es nicht weitere Termine beim selben Kunden am selben Tag gibt
                #Erweiterung: nur 1 Termin am gesamten Tag, analyse ergab, dass es bessere Ergebnisse liefert
                 if len([term1.kunde.name for term1 in tag.values()]) == 1:
                    info_array = {}
                    info_array['entfernung_tech1_original'] = termin['entfernung_original']
                    info_array['entfernung_tech2_original'] = entf2
                    info_array['entfernung_tech1_neu'] = entf1
                    info_array['entfernung_tech2_neu'] = termin['entfernung_potential']
                    info_array['einsparpotential'] = zu_sparen - (entf1 - entf2)
                    
                    info_array['tech1'] = techniker1
                    info_array['tech2'] = techniker2
                    
                    info_array['orig_tagplan'] = termin['orig_tagplan']
                    info_array['orig_tag'] = termin['orig_tag']
                    info_array['orig_stunde'] = termin['orig_stunde']
                    info_array['orig_termin'] = termin['termin']
                    
                    info_array['neu_tagplan'] = tag
                    info_array['neu_tag'] = i
                    info_array['neu_stunde'] = stunde
                    info_array['neu_termin'] = term

                    termine_array.append(info_array)
    return termine_array
    
def finde_tauschkandidaten(testproblem):
    '''
    Hilfsfunktion, die fuer ein Problem alle Tauschkandidaten identifiziert
    '''
    moegl = []
    #Aufgrund von Greedy Approach lassen sich hier nicht 2 direkt tauschen
    verbesserungs_termine = [direkt_verbesserbar(tech, testproblem) for tech in testproblem.techniker]
    #Schritt 1: finde in Zieltechs Termine, die weniger Zeit verlieren wuerden
    for verbesserung_tech in verbesserungs_termine:
        for verbesserung_termin in verbesserung_tech:
            curr_tech = verbesserung_termin['orig_tech']
            for tausche_mit in verbesserung_termin['neuer_tech']:
                moegl = finde_moeglichkeiten(curr_tech, tausche_mit, verbesserung_termin)
                #print len(moegl)
    return moegl

def tausche_termine(testproblem, test_backlog):
    sol_scores = [calc_solution_score(testproblem, test_backlog)]
    
    tauschkandidaten = finde_tauschkandidaten(testproblem)
    #nach einsparpotential sortieren
    sortierte_tauschkandidaten = []
    for t in tauschkandidaten:
        heapq.heappush(sortierte_tauschkandidaten, (-t['einsparpotential'], t))
    #Solange es noch Tauschkandidaten gibt, versuche zu verbessern
    while sortierte_tauschkandidaten:
        
        next_step = copy.deepcopy(testproblem)
        next_tausch = heapq.heappop(sortierte_tauschkandidaten)

        #erst techniker selektieren
        curr_tech = 0
        tausch_tech = 0
        for tech in next_step.techniker:
            if tech.name == next_tausch[1]['tech1'].name:
                curr_tech = tech
            if tech.name == next_tausch[1]['tech2'].name:
                tausch_tech = tech

        #Termine tauschen
        try:
            anfangszeit1 = einstellungen.zeiten_ab + next_tausch[1]['entfernung_tech1_neu']
            curr_tech.termine.kalender[next_tausch[1]['orig_tag']] = {}
            curr_tech.termine.kalender[next_tausch[1]['orig_tag']][anfangszeit1] = next_tausch[1]['neu_termin']

            anfangszeit2 = einstellungen.zeiten_ab + next_tausch[1]['entfernung_tech2_neu']
            tausch_tech.termine.kalender[next_tausch[1]['neu_tag']] = {}
            tausch_tech.termine.kalender[next_tausch[1]['neu_tag']][anfangszeit2] = next_tausch[1]['orig_termin']
        except:
            pass
        #dann Punkte
        curr_score = calc_solution_score(next_step, test_backlog)
        #Wenn Loesung besser ist, dann nimm sie als neue Hauptloesung
        if curr_score > sol_scores[-1]:
            testproblem = next_step
            sol_scores.append(curr_score)
            
    return sol_scores, testproblem
    
    
    
    
    
```

```python
#Berechne die Lösung mit unterschiedlich vielen Terminen
scores_simpel = []
scores_erweitert = []
scores_final = []
scores_local = []

for i in xrange(20, 350, 1):
    score_s = []
    score_e = []
    score_f = []
    score_l = []
    auswertungsproblem = problem.Problem(1, karte, 5, 20, i)
    auswertungsproblem, backlog_auswertung = solver.greedy_solver_final(auswertungsproblem)
    score_f.append(calc_solution_score(auswertungsproblem, backlog_auswertung))

    auswertungsproblem.reset_problem()
    auswertungsproblem, backlog_auswertung = solver.greedy_solver_erweitert(auswertungsproblem)
    score_e.append(calc_solution_score(auswertungsproblem, backlog_auswertung))

    auswertungsproblem.reset_problem()
    auswertungsproblem, backlog_auswertung = solver.greedy_solver(auswertungsproblem)
    score_s.append(calc_solution_score(auswertungsproblem, backlog_auswertung))
    
    auswertungsproblem.reset_problem()
    auswertungsproblem, backlog_auswertung = solver.greedy_heuristik(auswertungsproblem)
    sol_scores, auswertungsproblem = tausche_termine(auswertungsproblem, backlog_auswertung)
    score_l.append(sol_scores[-1])
    
    
    scores_simpel.append(np.mean(score_s))
    scores_erweitert.append(np.mean(score_e))
    scores_final.append(np.mean(score_f))
    scores_local.append(np.mean(score_l))
    
```
